# Find-S Algorithm 

- Find-S algorithm is a machine learning algorithm used for supervised learning. 
- It is used to find the most specific hypothesis that fits all the positive examples in the training data. 
- The algorithm starts with the most specific hypothesis and generalizes it as it encounters more positive examples.


### Python implementation of Find-S algorithm:


```python
# Define the training data
training_data = [
    ('Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same', True),
    ('Sunny', 'Warm', 'High', 'Strong', 'Warm', 'Same', True),
    ('Rainy', 'Cold', 'High', 'Strong', 'Warm', 'Change', False),
    ('Sunny', 'Warm', 'High', 'Strong', 'Cool', 'Change', True)
]
```

```python
# Initialize the hypothesis with the first positive example
hypothesis = ['0'] * 6
for instance in training_data:
    if instance[-1]:
        for i in range(6):
            if hypothesis[i] == '0':
                hypothesis[i] = instance[i]
            elif hypothesis[i] != instance[i]:
                hypothesis[i] = '?'

# Print the final hypothesis
print(hypothesis)
```

- In this example, we have four training examples with six attributes each.
- The last attribute is the target variable, which is either True or False. 
- We start with the most specific hypothesis, which is represented by '0' for each attribute. 
- We then iterate through the training examples and update the hypothesis for each positive example. If the attribute value is '0', we replace it with the corresponding value from the positive example. 
- If the attribute value is not '0' and does not match the corresponding value from the positive example, we replace it with '?' to indicate that it is unknown. Finally, we print the final hypothesis.

```python
import pandas as pd
import numpy as np
url="https://raw.githubusercontent.com/svkarthik86/Meachine_Learning/main/weather_dataset.csv"
#to read the data in the csv file
data = pd.read_csv(url)
print(data,"\n")
 
#making an array of all the attributes
d = np.array(data)[:,:-1]
print("\n The attributes are: ",d)
 
#segragating the target that has positive and negative examples
target = np.array(data)[:,-1]
print("\n The target is: ",target)
 
#training function to implement find-s algorithm
def train(c,t):
    for i, val in enumerate(t):
        if val == "Yes":
            specific_hypothesis = c[i].copy()
            break
             
    for i, val in enumerate(c):
        if t[i] == "Yes":
            for x in range(len(specific_hypothesis)):
                if val[x] != specific_hypothesis[x]:
                    specific_hypothesis[x] = '?'
                else:
                    pass
                 
    return specific_hypothesis
 
#obtaining the final hypothesis
print("\n The final hypothesis is:",train(d,target))
```
